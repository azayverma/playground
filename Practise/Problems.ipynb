Given an array of integers, find the pair of adjacent elements that has the largest product and return that product

```python
def get_largest_adjacent_product(arr):
  prod=arr[0]*arr[1]
  for i in range(0,len(arr)-1):
    prod=max(prod,arr[i]*arr[i+1])
  return prod
```

```python
import numpy as np
get_largest_adjacent_product(np.array([1,2,3,1])) # List in python is basically dynamic array. Arrays are more eficient and take less memory
```

```python
# O(1)
def swap(a,b):
  # a=a+b
  # b=a-b
  # a=a-b
  a,b=b,a
  return a,b
```

The problem is that we want to reverse a T[] array in O(N) linear time complexity and we want the algorithm to be in-place as well - so no additional memory can be used!

For example: input is [1,2,3,4,5] then the output is [5,4,3,2,1]

```python
def reverseArray(arr):
  # O(N/2)
  # n=len(arr)
  # for i in range(0,int(n/2)):
  #   arr[i],arr[n-1-i]=arr[n-1-i],arr[i]
  # return arr
  return arr[::-1] #O(N)
```

Our task is to design an efficient algorithm to reverse a given integer. For example if the input of the algorithm is 1234 then the output should be 4321.

```python
def reverse_int(num):
  #return int(str(num)[::-1]) O(N)
  # Below is also O(N)
  reversed_num=0
  while num>0:
    rem=num%10
    reversed_num=reversed_num*10+rem
    num=int(num/10)
  return reversed_num
```

```python
reverse_int(1234)
```

The problem is that we want to find duplicates in a one-dimensional array of integers in O(N) running time where the integer values are smaller than the length of the array!

For example: if we have a list [1, 2, 3, 1, 5] then the algorithm can detect that there are a duplicate with value 1.

Note: the array can not contain items smaller than 0 and items with values greater than the size of the list. This is how we can achieve O(N) linear running time complexity!

```python
def findDuplicates(p_list):
  # O(N^2)
  # dup_list=[]
  # for element in set(p_list):
  #   count_element=p_list.count(element)
  #   if count_element>1:
  #     dup_list.append(element)
  # return dup_list
  p_set=set(p_list) # O(N)
  count_dict={}
  dup_list=[]
  for element in p_set: # O(N)
    count_dict[element]=0
  for element in p_list: # O(N)
    count_dict[element]=count_dict[element]+1
    if count_dict[element]>1:
      dup_list.append(element) # O(1)
  return list(set(dup_list))
```

```python
findDuplicates([1, 2, 1, 2, 5])
```

```python
def findDuplicates_inplace(nums):
  dup_list=[]
  for num in nums:    
    if nums[abs(num)]>=0:
      nums[abs(num)]=-1*nums[abs(num)]
    else:      
      dup=abs(nums[abs(num)])
      dup_list.append(dup)
  return dup_list
```

```python
findDuplicates_inplace([1, 2, 1, 2,4])
```

```python
from datetime import datetime
balances = [1000, 1500]
requests = ["withdraw 1613327630 2 480", "withdraw 1613327644 2 800", "withdraw 1614105244 1 100", "deposit 1614108844 2 200", "withdraw 1614108845 2 150"]
def f(balances,requests):
  request_id=1
  for request in requests:
    request_split=request.split()
    transaction_type,txntimestamp,holder_id,amount=request_split
    holder_id,amount,txntimestamp=int(holder_id),float(amount),int(txntimestamp)
    if holder_id>len(balances):
      return [-request_id]
    if transaction_type=='withdraw':
      if balances[holder_id-1]>=amount:
        balances[holder_id-1]=balances[holder_id-1]-amount
        if (datetime.now()-datetime.fromtimestamp(txntimestamp)).days>=24:
          balances[holder_id-1]=balances[holder_id-1]+int(amount*0.02)
      else:
        return [-request_id]
    else:
      if transaction_type=='deposit':
        balances[holder_id-1]=balances[holder_id-1]+amount
    request_id+=1
    
  return balances
f(balances,requests)
```
