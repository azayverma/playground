```python
# Given a time, return the angle between hour clock and minute clock
def calcAngle(time):
  hourHand,minuteHand=[int(i) for i in time.split(":")]
  hourHand=hourHand%12

  # Minute Hand : Full Circle = 60 minutes for 360 degree => 1 minute=6 degree
  # Hour Hand : Full Circle = 12 hrs for 360 degrees => 1 hr = 30 degree
  # 60 minute = 1 hour = 1/12*360=30 degree of hour hand
  # 1 minute = 30/60=0.5 degree of hour

  minuteHand_angle=minuteHand*6
  hourHand_angle=hourHand*30+0.5*minuteHand
  return hourHand_angle-minuteHand_angle

calcAngle("3:15")
```

```python
# Find special product
import numpy as np
def findSpecialProduct(nums):
  # prod=np.prod(nums)
  # return [int(prod/no) for no in nums] #O(n) using division
  n=len(nums)
  return [np.prod(nums[i+1:]+nums[:i]) for i in range(n)] #O(n) without using division
  
findSpecialProduct([1,2,3,4,5])
```

```python
# Given a list of points on the 2-D plane and an integer K. The task is to find K closest points to the origin and print them.
def closest_points(points,k):
  points.sort(key=lambda i:i[0]**2+i[1]**2)
  return points[:k]

closest_points([[3, 3], [5, -1], [-2, 4]], 2)
```

```python
# Rain Trapping Problem
def rain_trapped(arr):
  res=0
  n=len(arr)
  for i in range(1,n-1):
    # Find the maximum element on its left inclusive
    left_max=max(arr[:i+1])
    # Find the maximum element on its right inclusive
    right_max=max(arr[i:])
    res = res + (min(left_max, right_max) - arr[i])
  
  return res
rain_trapped([0, 1, 0, 2, 1, 0, 1, 3, 2, 1, 2, 1])
```
